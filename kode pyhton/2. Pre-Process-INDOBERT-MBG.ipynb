```python
import pandas as pd
from googletrans import Translator
import time
import re

# Inisialisasi Translator
translator = Translator()

# Nama file yang akan diproses
file_path = 'dataset_full_cleaned.xlsx'

# Nama kolom yang berisi komentar yang akan diterjemahkan
kolom_komentar = 'comment_cleaned'

# Fungsi untuk menerjemahkan teks dari Bahasa Inggris ke Bahasa Indonesia
def translate_to_indonesian(text):
    # Pastikan teks yang akan diterjemahkan adalah string dan tidak kosong
    if pd.isna(text) or not isinstance(text, str) or len(text.strip()) == 0:
        return text

    try:
        # Pisahkan teks di dalam [...] dan di luar [...]
        pattern = r'(\[.*?\])'
        parts = re.split(pattern, text)

        translated_parts = []

        for part in parts:
            # Jika part adalah teks dalam [...], simpan tanpa terjemahan
            if part.startswith('[') and part.endswith(']'):
                translated_parts.append(part)
            # Jika bukan, terjemahkan part tersebut
            elif part.strip():  # Hanya terjemahkan jika tidak kosong
                # Terjemahkan langsung dari auto-detect ke Indonesia
                # Google Translate akan otomatis mendeteksi dan menerjemahkan
                # bagian Inggris, sementara membiarkan bagian Indonesia
                translated = translator.translate(part, src='auto', dest='id')
                translated_parts.append(translated.text)
                # Tambahkan delay kecil untuk menghindari rate limit
                time.sleep(0.2)
            else:
                translated_parts.append(part)

        result = ''.join(translated_parts)
        print(f"Original: '{text}'")
        print(f"Translated: '{result}'\n")
        return result

    except Exception as e:
        print(f"Error saat menerjemahkan '{text[:50]}...': {e}")
        # Kembalikan teks asli jika terjadi error
        return text

# --- Proses Utama ---
try:
    # Membaca data dari file Excel
    df = pd.read_excel(file_path)

    # Menambahkan kolom baru untuk menyimpan hasil terjemahan
    df['comment_translated'] = df[kolom_komentar].apply(translate_to_indonesian)

    # Simpan DataFrame yang telah diperbarui ke file Excel baru
    output_file_path = 'dataset_translated.xlsx'
    df.to_excel(output_file_path, index=False)

    print("\nProses terjemahan selesai!")
    print(f"File hasil tersimpan di: {output_file_path}")

except FileNotFoundError:
    print(f"Error: File '{file_path}' tidak ditemukan. Pastikan Anda sudah mengunggahnya ke Google Colab.")
except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

```python
from graphviz import Digraph

# Membuat diagram dengan konfigurasi yang lebih baik
dot = Digraph(comment='Arsitektur Sistem Analisis Sentimen IndoBERT',
              format='png', engine='dot')

# Pengaturan global diagram
dot.attr(rankdir='TB',  # Top to Bottom untuk alur yang lebih natural
         size='12,10',
         dpi='300',
         bgcolor='white',
         fontname='Arial',
         fontsize='12')

# Pengaturan default untuk nodes
dot.attr('node',
         shape='box',
         style='rounded,filled',
         fontname='Arial',
         fontsize='11',
         margin='0.3,0.2')

# Pengaturan default untuk edges
dot.attr('edge',
         fontname='Arial',
         fontsize='9',
         color='#2E8B57')

# === FASE 1: DATA COLLECTION ===
with dot.subgraph(name='cluster_0') as c:
    c.attr(style='filled', color='lightblue', label='Fase 1: Pengumpulan Data')
    c.node('A', 'Data Reddit\n(6000 komentar)\nraw text',
           fillcolor='#E6F3FF', color='#4A90E2')

# === FASE 2: DATA PREPROCESSING ===
with dot.subgraph(name='cluster_1') as c:
    c.attr(style='filled', color='lightgreen', label='Fase 2: Preprocessing')
    c.node('B', 'Text Preprocessing\n• Cleaning\n• Normalisasi\n• Translasi ke Bahasa Indonesia',
           fillcolor='#E6FFE6', color='#5CB85C')

# === FASE 3: DATA PREPARATION ===
with dot.subgraph(name='cluster_2') as c:
    c.attr(style='filled', color='lightyellow', label='Fase 3: Persiapan Data')
    c.node('C', 'Manual Labeling\n(3000 data)\nPositif/Negatif/Netral',
           fillcolor='#FFFACD', color='#F0AD4E')
    c.node('D', 'Data Splitting\n• Training: 2400\n• Testing: 600',
           fillcolor='#FFFACD', color='#F0AD4E')

# === FASE 4: MODEL TRAINING ===
with dot.subgraph(name='cluster_3') as c:
    c.attr(style='filled', color='lightcoral', label='Fase 4: Training Model')
    c.node('E', 'IndoBERT\nFine-tuning\n• Hyperparameter tuning\n• Optimization',
           fillcolor='#FFE4E1', color='#D9534F')

# === FASE 5: EVALUATION ===
with dot.subgraph(name='cluster_4') as c:
    c.attr(style='filled', color='plum', label='Fase 5: Evaluasi')
    c.node('F', 'Model Evaluation\n• Akurasi\n• Presisi\n• Recall\n• F1-Score',
           fillcolor='#F0E6FF', color='#8E44AD')

# === FASE 6: PREDICTION & ANALYSIS ===
with dot.subgraph(name='cluster_5') as c:
    c.attr(style='filled', color='lightsteelblue', label='Fase 6: Prediksi & Analisis')
    c.node('G', 'Prediksi Sentimen\n(6000 data lengkap)',
           fillcolor='#E6F0FF', color='#5DADE2')
    c.node('H', 'Analisis Hasil\n• Distribusi sentimen\n• WordCloud\n• Visualisasi',
           fillcolor='#E6F0FF', color='#5DADE2')

# === FASE 7: OUTPUT ===
with dot.subgraph(name='cluster_6') as c:
    c.attr(style='filled', color='lightgray', label='Output')
    c.node('I', 'Laporan Akhir\n• Model performance\n• Insight sentimen\n• Rekomendasi',
           fillcolor='#F5F5F5', color='#6C757D')

# Koneksi utama
dot.edges(['AB', 'BC', 'CD', 'DE', 'EF'])

# Branching untuk prediksi
dot.edge('F', 'G', label='Model OK?')
dot.edge('G', 'H')
dot.edge('H', 'I')

# Feedback loop jika model perlu improvement
dot.edge('F', 'E', label='perlu improvement', style='dashed', color='red')

# Tambahan informasi
dot.node('legend', '''Legend:
- Biru: Data Collection
- Hijau: Preprocessing
- Kuning: Data Preparation
- Merah: Model Training
- Ungu: Evaluation
- Biru Muda: Prediction
- Abu: Output''',
         shape='note',
         fillcolor='white',
         fontsize='9')

# Render diagram
dot.render('arsitektur_sistem_improved',
           view=True,
           cleanup=True)

print("Diagram arsitektur sistem berhasil dibuat!")
print("File output: arsitektur_sistem_improved.png")
```
